```python
data = pd.read_csv("IDEA-FRM_filtered_tenders_with_metadata_geocoded_DISTRICTS - IDEA-FRM_filtered_tenders_with_metadata_geocoded_revcir(1).csv")
```

```python
roads_bridges_array = ["Roads","Bridges","Road","Bridge","Storm Water Drain","Drain","Box Cul",
                       "Box Culvert","Box Culv","culvert",'Embankment','Embkt','River Bank Protection',
                       'Bank Protection','Dyke','Dyke Wall','Dyke Walls','Silt','Siltation','Sluice','Breach',
                       "Bund","Bunds","Bundh"]
erosion_array = ["Anti Erosion","AE","A/E","A E","Erosion","Eroded","Erroded"
,"Anti Errosion"
,"Errosion"
]
im_array = ["IM","Immediate","Emergency","Relief","GR","G.R.","I.M.","SDRF","Immediate Measure"]
new_constructions_array = ["Construction","Const.","Const. ","Contruction"]
repair_array = ["Improvemen"
,"Imp. "
,"Impvt"
,"Impt. "
,"Impt."
,"Repair"
,"Restoration"
,"Reconstruction"
,"Reconstn"
,"Improvement"
,"Repairing"
,"Recoupment"
,"Raising and Strengthening"
,"R/S"
,"Reconstnn"
,"M and R"
,"MnR"
,"M-R"
,"M n R"
,"Upgradation"
,"Renovation"
,"Reparing/Renovation"
,"Up-gradation"
,"Strengthening"
,"M-R "
,"MR"
,"Widening"
,"R S"
,"R/S "
,"Extension"
,"Replacement"
,"Up-gradation "
,"Re-Shaping"
,"Re-Grading"
]
relief_preparation_array = ["Shelter","Shelter","Shelters","Rehabilitation","Tarpaulin"]
```

```python
data['Roads,Bridges and Embankments'] = False
data['Erosion'] = False
data['Immediate Measure'] = False
data['New Constructions'] = False
data['Repair,Restoration and Improvement'] = False
data['Relief and Preparation'] = False
data['Goods'] = False
```

```python
def product_category(data,roads_bridges_array,erosion_array,im_array,new_constructions_array,repair_array,relief_preparation_array):
    for index, row in data.iterrows():
        tender_slug = str(row['tender_externalreference']) + ' ' + str(row['tender_title']) + ' ' + str(row['Work_Description'])
        tender_slug = re.sub('[^a-zA-Z0-9 \n\.]', ' ', tender_slug)
        for keyword_ero in erosion_array:
            keyword_count_ero = len(re.findall(r"\b%s\b" % keyword_ero.lower(), tender_slug.lower()))
            if keyword_count_ero >= 1:
                data.loc[index,"Erosion"] = True
        for keyword_roads in roads_bridges_array:
            keyword_count_roads = len(re.findall(r"\b%s\b" % keyword_roads.lower(), tender_slug.lower()))
            if keyword_count_roads >= 1 and data.loc[index,"Erosion"] == False:
                data.loc[index,"Roads,Bridges and Embankments"] = True
        for keyword_im in im_array:
            keyword_count_im = len(re.findall(r"\b%s\b" % keyword_im.lower(), tender_slug.lower()))
            if keyword_count_im >= 1:
                data.loc[index,"Immediate Measure"] = True
        for keyword_im in new_constructions_array:
            keyword_count_im = len(re.findall(r"\b%s\b" % keyword_im.lower(), tender_slug.lower()))
            if keyword_count_im >= 1:
                data.loc[index,"New Constructions"] = True
        for keyword_im in repair_array:
            keyword_count_im = len(re.findall(r"\b%s\b" % keyword_im.lower(), tender_slug.lower()))
            if keyword_count_im >= 1:
                data.loc[index,"Repair,Restoration and Improvement"] = True
        for keyword_im in relief_preparation_array:
            keyword_count_im = len(re.findall(r"\b%s\b" % keyword_im.lower(), tender_slug.lower()))
            if keyword_count_im >= 1:
                data.loc[index,"Relief and Preparation"] = True
    return data
```

```python
data_new = product_category(data,roads_bridges_array,erosion_array,im_array,new_constructions_array,repair_array,relief_preparation_array)
```
