```python
import re
from datetime import datetime
```

```python
# Renaming rows
hp_tender = hp_tender.rename(columns={'Tender Reference Number':'tender_externalreference', 'Title':'tender_title'})
hp_tender
```

```python
# Updated function to handle non-string values gracefully
def extract_date(text):
    if not isinstance(text, str):
        return None
    date_patterns = [
        r'\b(\d{2})[-/](\d{2})[-/](\d{2,4})\b',       # dd-mm-yy, dd-mm-yyyy, dd/mm/yyyy
        r'\b(\d{2})[.](\d{2})[.](\d{4})\b',           # dd.mm.yyyy
        r'\b(\d{2})(\d{2})(\d{4})\b'                  # ddmmyyyy
    ]
    for pattern in date_patterns:
        match = re.search(pattern, text)
        if match:
            date_str = match.group(0)
            try:
                if len(match.group(3)) == 2:
                    date_format = "%d-%m-%y"
                elif '-' in date_str or '/' in date_str:
                    date_format = "%d-%m-%Y" if '-' in date_str else "%d/%m/%Y"
                elif '.' in date_str:
                    date_format = "%d.%m.%Y"
                else:
                    date_format = "%d%m%Y"
                return datetime.strptime(date_str, date_format)
            except ValueError:
                continue
    return None

# Apply the function to extract dates and create a new column
hp_tender['Published Date'] = hp_tender['tender_externalreference'].apply(extract_date)
hp_tender['month'] = hp_tender['Published Date'].dt.strftime('%Y_%m')


# Drop rows where the date could not be extracted
df_cleaned = hp_tender.dropna(subset=['Published Date','month'])


# Save the cleaned DataFrame to a new file
cleaned_file_path = r'D:\CivicDataLabs_IDS-DRR\IDS-DRR_Github\HP temp\flood-data-ecosystem-Himachal-Pradesh\Sources\TENDERS\data\monthly_tenders\HP_Tenders_Cleaned.csv'
df_cleaned.to_csv(cleaned_file_path, index=False)

# Display the cleaned DataFrame and the saved file path
df_cleaned.head(), cleaned_file_path
```
